```python
sns.displot(df["Age"])
```

```python
sns.boxplot(df["Age"]) # From box plot we can observe that there are no value beyond 1.5*IQR, hence there is no outliers for variable age.
```

```python
# To identify outliers 
# IQR method
Q1=df["Age"].quantile(0.25)
Q3=df["Age"].quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
```

```python
upper_limit_age= Q3+(1.5*IQR)
lower_limit_age= Q1-(1.5*IQR)
print(upper_limit_age, lower_limit_age)
```

```python
#Idntification of outliers
# trim data based on 1.5*IQR
new_df=df.loc[(df["Age"]<upper_limit_age) & (df["Age"]>lower_limit_age) ]
print("before removing outliers:", len(df))
print("after removing outliers:", len(new_df))
print("No. of outliers:", len(df)-len(new_df))
#There is no outlier for variable "Age"
```

```python
sns.displot(df["Income"]) # Histogram Shows positely skewed distribution. 
```

```python
sns.boxplot(df["Income"])
# Many values beyound Q3+1.5 *IQR indicates presence of outliers. 
```

```python
# IQR method
Q1=df["Income"].quantile(0.25)
Q3=df["Income"].quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
upper_limit_Income= Q3+(1.5*IQR)
lower_limit_Income= Q1-(1.5*IQR)
print(upper_limit_Income, lower_limit_Income)
```

```python
# trim data based on 1.5*IQR
new_df=df.loc[(df["Income"]<upper_limit_Income) & (df["Income"]>lower_limit_Income) ]
print("before removing outliers:", len(df))
print("after removing outliers:", len(new_df))
print("No. of outliers:", len(df)-len(new_df))
```

```python
#Capping method: replacing ouliers by upper or lower limit
new_df=df.copy()
new_df.loc[(new_df["Income"]>upper_limit_Income), "Income"]= upper_limit_Income
new_df.loc[(new_df["Income"]<lower_limit_Income), "Income"]= lower_limit_Income
            
```

```python
#Box plot after using capping method. which indicates no presence of outliers
sns.boxplot(new_df["Income"])
```
